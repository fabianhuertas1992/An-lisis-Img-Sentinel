# Análisis NDVI

```python
# @title Análisis total Imágenes Sentinel Dos periodos (**Correciones NDVI + Nuevas coberturas NDVI Adicionales**)

# ===========================================
# 1) IMPORTACIONES Y CONSTANTES GLOBALES
# ===========================================

import os
import numpy as np
import pandas as pd
import rasterio
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import re
from matplotlib.patches import Patch
import plotly.graph_objs as go
import plotly.offline as pyo
from urllib.request import urlopen
import json
from shapely.geometry import shape
import geopandas as gpd
import matplotlib.patches as mpatches

# Constantes globales
PIXEL_SIZE = 10          # Tamaño del píxel en metros (10 m x 10 m)
HECTARE_CONVERSION = 10000  # Conversión de m² a hectáreas

# ------------------------------
#  Cargar GeoJSON de Colombia
# ------------------------------
with urlopen('https://gist.githubusercontent.com/john-guerra/43c7656821069d00dcbc/raw/be6a6e239cd5b5b803c6e7c2ec405b793a9064dd/Colombia.geo.json') as response:
    counties = json.load(response)

features = counties['features']
geoms = [shape(f['geometry']) for f in features]
props = [f['properties'] for f in features]
departamentos = gpd.GeoDataFrame(props, geometry=geoms, crs="EPSG:4326")

# ------------------------------
# Cargar tu polígono de trabajo
# ------------------------------
#GEOJSON_PATH_MANUAL = "ruta/a/tu/archivo.geojson"  # Asegúrate de definir esta variable
region_gdf = gpd.read_file(GEOJSON_PATH_MANUAL)

# ------------------------------
# Detectar el departamento que contiene el polígono
# ------------------------------
region_union = region_gdf.unary_union
departamento_encontrado = departamentos[departamentos.contains(region_union)]

if not departamento_encontrado.empty:
    nombre_dpto = departamento_encontrado.iloc[0].get("NOMBRE_DPT", "Depto Detectado")
    print(f"📍 Departamento encontrado: {nombre_dpto}")
else:
    print("⚠️ No se pudo encontrar el departamento")
    nombre_dpto = "Desconocido"




# ===========================================
# 2) FUNCIONES DE UTILIDAD
# ===========================================

def buscar_archivo_por_banda(carpeta, banda):
    """
    Busca un archivo en la carpeta que contenga la banda especificada ('B04' o 'B08').
    Devuelve la ruta completa del archivo encontrado.
    """
    for archivo in os.listdir(carpeta):
        if banda in archivo:
            return os.path.join(carpeta, archivo)
    return None

'''def extraer_prefijo_archivo(ruta_archivo):
    """
    Dada la ruta de un archivo con nombre como:
       'Cauca_0.0_2019-01-01_Cajibio_Sentinel-2_B04.tiff'
    Retorna algo como:
       'Cauca_0.0_'
    Ajusta la lógica si tus archivos tienen un formato distinto.
    """
    nombre_archivo = os.path.basename(ruta_archivo)
    partes = nombre_archivo.split('_')
    if len(partes) >= 2:
        # Tomamos las dos primeras partes y les sumamos el guion bajo final
        prefijo = f"{partes[0]}_{partes[1]}_"
    else:
        # Si no hay suficientes partes, usamos todo el nombre
        prefijo = nombre_archivo
    return prefijo'''

# Extraer el nombre base del JSON sin la extensión
def extraer_nombre_json(ruta_json):
    """
    Extrae el nombre base del archivo JSON sin la extensión.
    Por ejemplo, de "Tecnicafe_Cauca.json" devuelve "Tecnicafe_Cauca"
    """
    nombre_archivo = os.path.basename(ruta_json)
    nombre_base = os.path.splitext(nombre_archivo)[0]  # Quita la extensión
    return nombre_base

def extraer_fecha_desde_ruta(ruta):
    """
    Extrae una fecha en formato YYYY-MM-DD desde una ruta como:
    /.../Tecnicafe_Cauca_20190830 → 2019-08-30
    """
    patron = r'(\d{8})$'
    coincidencia = re.search(patron, os.path.basename(os.path.dirname(ruta)))
    if coincidencia:
        fecha_cruda = coincidencia.group(1)
        return f"{fecha_cruda[:4]}-{fecha_cruda[4:6]}-{fecha_cruda[6:]}"
    return "FechaDesconocida"

def process_image(bandas):
    """
    Procesa las imágenes NIR y ROJA para calcular el NDVI con normalización
    (cada banda se divide por su máximo valor).
    """
    try:
        with rasterio.open(bandas['nir']) as src:
            banda_NIR = src.read(1).astype('float32')
        with rasterio.open(bandas['red']) as src:
            banda_roja = src.read(1).astype('float32')

        # Crear máscara válida inicial
        mascara = (banda_roja > 0) & (banda_NIR > 0)

        # Normalización (división por el valor máximo de cada banda)
        norm_NIR = banda_NIR / np.max(banda_NIR[mascara])
        norm_RED = banda_roja / np.max(banda_roja[mascara])

        # Aplicar máscara después de la normalización
        norm_NIR_masked = np.ma.array(norm_NIR, mask=~mascara)
        norm_RED_masked = np.ma.array(norm_RED, mask=~mascara)

        # Cálculo del NDVI usando las bandas normalizadas
        ndvi = (norm_NIR_masked - norm_RED_masked) / (norm_NIR_masked + norm_RED_masked)

        return ndvi, mascara

    except Exception as e:
        print(f"Error al procesar imágenes: {e}")
        return None, None

def calcular_area_hectareas(mascara):
    """
    Devuelve el área total (en ha) de píxeles válidos en la máscara.
    """
    area_m2 = np.sum(mascara) * (PIXEL_SIZE ** 2)
    return area_m2 / HECTARE_CONVERSION

def extraer_año(ruta_archivo):
    """
    Extrae el año de un nombre de archivo con patrón (YYYY-MM-DD).
    """
    nombre_archivo = os.path.basename(ruta_archivo)
    patron = r'(\d{4})-\d{2}-\d{2}'
    coincidencia = re.search(patron, nombre_archivo)
    if coincidencia:
        return coincidencia.group(1)
    return "Desconocido"

def validar_mascara_comun(mascara_comun):
    """
    Muestra cuántos píxeles válidos existen en la máscara común.
    """
    valid_pixels = np.sum(mascara_comun)
    print(f"Píxeles válidos en la máscara común: {valid_pixels}")
    if valid_pixels == 0:
        print("Advertencia: No se encontraron píxeles válidos. Revisa tus archivos de imagen.")

def generar_dataframe_estadisticas(ndvi, mascara, nombre_archivo, año):
    """
    Devuelve un diccionario con estadísticas clave de NDVI:
      - Total Área (ha)
      - NDVI Máximo
      - NDVI Mínimo
      - NDVI Promedio
    """
    area_total = calcular_area_hectareas(mascara)
    ndvi_max = np.nanmax(ndvi)
    ndvi_min = np.nanmin(ndvi)
    ndvi_promedio = np.nanmean(ndvi)
    return {
        "Año de la Imagen": año,
        "Total Área (ha)": area_total,
        "NDVI Máximo": ndvi_max,
        "NDVI Mínimo": ndvi_min,
        "NDVI Promedio": ndvi_promedio
    }

def generar_dataframe_resumen(ndvi_anterior, ndvi_posterior, mascara_anterior, mascara_posterior, grupos, etiquetas):
    def calcular_area(ndvi, mascara):
        conteos, _ = np.histogram(ndvi[mascara], bins=grupos)
        areas_ha = conteos * (PIXEL_SIZE**2) / HECTARE_CONVERSION
        return areas_ha

    # Áreas anteriores y posteriores
    areas_anterior = calcular_area(ndvi_anterior, mascara_anterior)
    areas_posterior = calcular_area(ndvi_posterior, mascara_posterior)

    # Cálculo de porcentajes
    porcentaje_anterior = (areas_anterior / areas_anterior.sum()) * 100
    porcentaje_posterior = (areas_posterior / areas_posterior.sum()) * 100

    # Cálculo del cambio (hectáreas y porcentaje)
    cambio_area = areas_posterior - areas_anterior
    cambio_porcentaje = porcentaje_posterior - porcentaje_anterior

    # Construcción del DataFrame resumen
    df_resumen = pd.DataFrame({
        'Categoría NDVI': etiquetas,
        'Área Anterior (ha)': areas_anterior,
        'Porcentaje Anterior (%)': porcentaje_anterior,
        'Área Posterior (ha)': areas_posterior,
        'Porcentaje Posterior (%)': porcentaje_posterior,
        'Cambio de Área (ha)': cambio_area,
        'Cambio en Porcentaje (%)': cambio_porcentaje
    })

    return df_resumen

# ===========================================
# 3) MAIN: CARGA DE DATOS Y PROCESO NDVI
# ===========================================

# Buscar archivos por bandas
nir_anterior = buscar_archivo_por_banda(carpeta_anterior, "B08")
red_anterior = buscar_archivo_por_banda(carpeta_anterior, "B04")
nir_posterior = buscar_archivo_por_banda(carpeta_posterior, "B08")
red_posterior = buscar_archivo_por_banda(carpeta_posterior, "B04")

print("NIR Anterior:", nir_anterior)
print("Red Anterior:", red_anterior)
print("NIR Posterior:", nir_posterior)
print("Red Posterior:", red_posterior)

# Extraer prefijo para nombrar tus archivos de salida
'''prefijo_archivo = extraer_prefijo_archivo(nir_anterior)
print("Prefijo Detectado:", prefijo_archivo)'''

prefijo_archivo = extraer_nombre_json(GEOJSON_PATH_MANUAL) + "_"
print("Prefijo JSON Detectado:", prefijo_archivo)

# Extraer años para usarlos en las gráficas
año_anterior = extraer_fecha_desde_ruta(nir_anterior)
año_posterior = extraer_fecha_desde_ruta(nir_posterior)

# Procesar imágenes y calcular NDVI
ndvi_anterior, mascara_anterior = process_image({'nir': nir_anterior, 'red': red_anterior})
ndvi_posterior, mascara_posterior = process_image({'nir': nir_posterior, 'red': red_posterior})

# Crear máscara común
mascara_comun = mascara_anterior & mascara_posterior
validar_mascara_comun(mascara_comun)

# Aplicar máscara común
ndvi_anterior = np.ma.array(ndvi_anterior, mask=~mascara_comun)
ndvi_posterior = np.ma.array(ndvi_posterior, mask=~mascara_comun)

# Calcular diferencia NDVI
diferencia_ndvi = ndvi_posterior - ndvi_anterior
mascara_diferencia = mascara_anterior & mascara_posterior  # Idéntica a mascara_comun

# ===========================================
# 4) ESTADÍSTICAS Y DATAFRAMES
# ===========================================

# Generar datos estadísticos
datos = []
datos.append(generar_dataframe_estadisticas(ndvi_anterior, mascara_anterior, nir_anterior, f"NDVI - {año_anterior}"))
datos.append(generar_dataframe_estadisticas(ndvi_posterior, mascara_posterior, nir_posterior, f"NDVI - {año_posterior}"))
datos.append(generar_dataframe_estadisticas(diferencia_ndvi, mascara_diferencia, "Diferencia NDVI", f"Diferencia - {año_anterior} - {año_posterior}"))
df_estadisticas = pd.DataFrame(datos)

# Aplanar datos para comparaciones
ndvi_anterior_flat = ndvi_anterior.compressed()
ndvi_posterior_flat = ndvi_posterior.compressed()

# Crear un rango de bins común (8 grupos)
min_value = min(np.min(ndvi_anterior_flat), np.min(ndvi_posterior_flat))
max_value = max(np.max(ndvi_anterior_flat), np.max(ndvi_posterior_flat))
#grupos = np.linspace(min_value, max_value, 9). #Esto si quiero dejar los grupos de manera automatica
#etiquetas = [f'{grupos[i]:.2f} - {grupos[i+1]:.2f}' for i in range(len(grupos)-1)]
# 1) Nuevas definiciones globales:
#grupos = [ -1.0, 0.01, 0.1, 0.2, 0.4, 1.0]
grupos = [  -1.0, 0.01, 0.2, 0.4, 0.8, 1.0]
etiquetas = [
    'Otros (<0.01)',
    'Suelo sin vegetación (0.01 - 0.2)',
    'Vegetación ligera (0.2 - 0.2)',
    'Vegetación mediana (0.4 - 0.8)',
    'Vegetación alta (>0.8)'
]
#colores = ['#0e2f6f', '#8a4c04', '#fdff5b', '#59cf52', '#386602', '#323e01']
colores = ['#24427c', '#df7d28', '#faea73', '#64bc6a', '#246128']

ndvi_cmap = mcolors.ListedColormap(colores)
norm = mcolors.BoundaryNorm(grupos, ndvi_cmap.N)

# Calcular conteos en cada grupo
conteo_anterior, _ = np.histogram(ndvi_anterior_flat, bins=grupos)
conteo_posterior, _ = np.histogram(ndvi_posterior_flat, bins=grupos)

df_comparativo = pd.DataFrame({
    'Grupo NDVI': etiquetas,
    f'Conteo NDVI Anterior ({año_anterior})': conteo_anterior,
    f'Conteo NDVI Posterior ({año_posterior})': conteo_posterior
})

# ===========================================
# 5) GENERACIÓN DE GRÁFICOS
# ===========================================

# GRÁFICO NDVI NORMAL (Anterior, Posterior, Diferencia) ---

def generar_grafico_ndvi(ndvi_anterior, ndvi_posterior, año_anterior, año_posterior,
                         mascara_anterior, mascara_posterior, prefijo_archivo):
    """
    Genera un gráfico con 3 paneles:
    - NDVI Anterior
    - NDVI Posterior
    - Diferencia NDVI
    Con las estadísticas básicas mostradas en cada panel.
    """
    diferencia_ndvi = ndvi_posterior - ndvi_anterior

    # Crear la figura con 3 paneles
    fig, axes = plt.subplots(1, 3, figsize=(22, 6))

    # Normalización para gráficos de NDVI
    norm = mcolors.Normalize(vmin=min_value, vmax=max_value)  # Ajusta según tu rango real
    norm_diff = mcolors.Normalize(vmin=-0.4, vmax=0.4)  # Ajusta según tu rango real

    # Quitar ejes y bordes
    for ax in axes:
        ax.set_xticks([])
        ax.set_yticks([])
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.spines['left'].set_visible(False)
        ax.spines['bottom'].set_visible(False)

    # Panel 1: NDVI Anterior
    img1 = axes[0].imshow(ndvi_anterior, cmap='RdYlGn', norm=norm)
    axes[0].set_title(f'NDVI  - {año_anterior}', fontsize=14, fontweight='bold')
    plt.colorbar(img1, ax=axes[0], orientation='horizontal', fraction=0.046, pad=0.13)

    # Estadísticos NDVI Anterior
    axes[0].text(
        0.5, -0.04,
        f'Total Área: {calcular_area_hectareas(mascara_anterior):.2f} ha | NDVI Máx: {np.nanmax(ndvi_anterior):.2f}\n'
        f'Promedio: {np.nanmean(ndvi_anterior):.2f} | NDVI Mín: {np.nanmin(ndvi_anterior):.2f}',
        transform=axes[0].transAxes, fontsize=12, ha='center', va='top',
        bbox=dict(facecolor='white', alpha=0.8)
    )

    # Panel 2: NDVI Posterior
    img2 = axes[1].imshow(ndvi_posterior, cmap= 'RdYlGn', norm=norm)
    axes[1].set_title(f'NDVI - {año_posterior}', fontsize=14, fontweight='bold')
    plt.colorbar(img2, ax=axes[1], orientation='horizontal', fraction=0.046, pad=0.13)

    # Estadísticos NDVI Posterior
    axes[1].text(
        0.5, -0.04,
        f'Total Área: {calcular_area_hectareas(mascara_posterior):.2f} ha | NDVI Máx: {np.nanmax(ndvi_posterior):.2f}\n'
        f'Promedio: {np.nanmean(ndvi_posterior):.2f} | NDVI Mín: {np.nanmin(ndvi_posterior):.2f}',
        transform=axes[1].transAxes, fontsize=12, ha='center', va='top',
        bbox=dict(facecolor='white', alpha=0.8)
    )

    # Panel 3: Diferencia NDVI
    img3 = axes[2].imshow(diferencia_ndvi, cmap='RdYlGn', norm=norm_diff)
    axes[2].set_title(f'Diferencia NDVI ({año_anterior} - {año_posterior})', fontsize=14, fontweight='bold')
    plt.colorbar(img3, ax=axes[2], orientation='horizontal', fraction=0.046, pad=0.13)

    # Estadísticos de Diferencia
    pos_change = (np.sum(diferencia_ndvi > 0) * (PIXEL_SIZE ** 2)) / HECTARE_CONVERSION
    neg_change = (np.sum(diferencia_ndvi < 0) * (PIXEL_SIZE ** 2)) / HECTARE_CONVERSION
    axes[2].text(
        0.5, -0.04,
        f'Área Total: {calcular_area_hectareas(mascara_anterior & mascara_posterior):.2f} ha | Cambio Pos: {pos_change:.2f} ha\n'
        f'Promedio Cambio: {np.nanmean(diferencia_ndvi):.2f} | Cambio Neg: {neg_change:.2f} ha',
        transform=axes[2].transAxes, fontsize=12, ha='center', va='top',
        bbox=dict(facecolor='white', alpha=0.8)
    )

    # Guardar la figura con el prefijo
    nombre_archivo = os.path.join(
        ruta_guardado,
        f'{prefijo_archivo}Analisis_NDVI_{año_anterior}_vs_{año_posterior}.png'
    )
    plt.tight_layout()
    plt.savefig(nombre_archivo, dpi=300, bbox_inches='tight')
    plt.show()
    print(f"Imagen guardada en: {nombre_archivo}")

# GRÁFICO NDVI CATEGORIZADO (Clasificación Estandarizada) ---

def generar_grafico_ndvi_categorizado(ndvi_anterior, ndvi_posterior, año_anterior, año_posterior,
                                      diferencia_ndvi, mascara_anterior, mascara_posterior, grupos, etiquetas, prefijo_archivo):
    fig, axes = plt.subplots(1, 3, figsize=(24, 6))
    for ax in axes:
        ax.axis('off')

    colores = ['#24427c', '#df7d28', '#faea73', '#64bc6a', '#246128']
    ndvi_cmap = mcolors.ListedColormap(colores)
    norm = mcolors.BoundaryNorm(grupos, ndvi_cmap.N)

    def calcular_area_por_clase(ndvi, mascara):
        ndvi_flat = ndvi[mascara]
        conteos, _ = np.histogram(ndvi_flat, bins=grupos)
        areas_ha = conteos * (PIXEL_SIZE**2) / HECTARE_CONVERSION
        area_total = np.sum(areas_ha)
        porcentajes = (areas_ha / area_total) * 100
        df_resumen = pd.DataFrame({
            'Categoría NDVI': etiquetas,
            'Área (ha)': areas_ha,
            'Porcentaje (%)': porcentajes
        })
        return df_resumen

    # NDVI anterior
    im1 = axes[0].imshow(ndvi_anterior, cmap=ndvi_cmap, norm=norm)
    axes[0].set_title(f'NDVI - {año_anterior}', fontsize=14, fontweight='bold')
    plt.colorbar(im1, ax=axes[0], orientation='horizontal', fraction=0.046, pad=0.01)
    df_area_anterior = calcular_area_por_clase(ndvi_anterior, mascara_anterior)
    resumen_texto_anterior = '\n'.join(df_area_anterior.apply(lambda row: f"{row['Categoría NDVI']}: {row['Área (ha)']:.2f} ha ({row['Porcentaje (%)']:.1f}%)", axis=1))
    axes[0].text(0.5, -0.12, resumen_texto_anterior, transform=axes[0].transAxes, fontsize=10, ha='center', va='top', bbox=dict(facecolor='white', alpha=0.8))

    # NDVI posterior
    im2 = axes[1].imshow(ndvi_posterior, cmap=ndvi_cmap, norm=norm)
    axes[1].set_title(f'NDVI - {año_posterior}', fontsize=14, fontweight='bold')
    plt.colorbar(im2, ax=axes[1], orientation='horizontal', fraction=0.046, pad=0.01)
    df_area_posterior = calcular_area_por_clase(ndvi_posterior, mascara_posterior)
    resumen_texto_posterior = '\n'.join(df_area_posterior.apply(lambda row: f"{row['Categoría NDVI']}: {row['Área (ha)']:.2f} ha ({row['Porcentaje (%)']:.1f}%)", axis=1))
    axes[1].text(0.5, -0.12, resumen_texto_posterior, transform=axes[1].transAxes, fontsize=10, ha='center', va='top', bbox=dict(facecolor='white', alpha=0.8))

    # Diferencia NDVI
    cmap_diff = plt.get_cmap(ndvi_cmap)
    norm_diff = mcolors.TwoSlopeNorm(vmin=-0.5, vcenter=0, vmax=0.5)
    im3 = axes[2].imshow(diferencia_ndvi, cmap=cmap_diff, norm=norm_diff)
    axes[2].set_title(f'Diferencia NDVI ({año_anterior} - {año_posterior})', fontsize=14, fontweight='bold')
    plt.colorbar(im3, ax=axes[2], orientation='horizontal', fraction=0.046, pad=0.01)

    df_area_anterior.set_index('Categoría NDVI', inplace=True)
    df_area_posterior.set_index('Categoría NDVI', inplace=True)
    cambio_area = df_area_posterior['Área (ha)'] - df_area_anterior['Área (ha)']
    cambio_texto = '\n'.join([f"{cat}: {cambio_area.loc[cat]:.2f} ha" for cat in etiquetas])
    axes[2].text(0.5, -0.12, cambio_texto, transform=axes[2].transAxes, fontsize=10, ha='center', va='top', bbox=dict(facecolor='white', alpha=0.8))

    patches = [Patch(color=colores[i], label=etiquetas[i]) for i in range(len(etiquetas))]
    fig.legend(handles=patches, loc='lower center', ncol=5, fontsize=10,
               title='Categorías NDVI', title_fontsize=12, bbox_to_anchor=(0.5, -0.10))

    nombre_archivo = os.path.join(
        ruta_guardado,
        f'{prefijo_archivo}Analisis_NDVI_Categorizado_{año_anterior}_vs_{año_posterior}.png'
    )
    plt.tight_layout()
    plt.savefig(nombre_archivo, dpi=300, bbox_inches='tight')
    plt.show()
    print(f"Imagen guardada en: {nombre_archivo}")


# GRÁFICO COMPARATIVO (Histograma vs. Barras) ---

def generar_graficos_comparativos(ndvi_anterior, ndvi_posterior, año_anterior, año_posterior, grupos, etiquetas, prefijo_archivo):
    ndvi_anterior_flat = ndvi_anterior.compressed()
    ndvi_posterior_flat = ndvi_posterior.compressed()

    conteo_anterior, _ = np.histogram(ndvi_anterior_flat, bins=grupos)
    conteo_posterior, _ = np.histogram(ndvi_posterior_flat, bins=grupos)

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(24, 6))

    bins = np.linspace(min_value, max_value) #-1.0, 1.0, 50)
    ax1.hist(ndvi_posterior_flat, bins=bins, alpha=0.6, label=f'NDVI Posterior - {año_posterior}',
             color='#d7191c', edgecolor='black')
    ax1.hist(ndvi_anterior_flat, bins=bins, alpha=0.6, label=f'NDVI Anterior - {año_anterior}',
             color='#2b83ba', edgecolor='black')
    ax1.set_title('Comparación de Histogramas NDVI', fontsize=16, fontweight='bold')
    ax1.set_xlabel('Valor NDVI', fontsize=14)
    ax1.set_ylabel('Frecuencia (Cantidad de píxeles)', fontsize=14)
    ax1.legend()

    x = np.arange(len(etiquetas))
    width = 0.4

    barras1 = ax2.bar(x - width/2, conteo_anterior, width, label=f'NDVI Anterior - {año_anterior}', color='#2b83ba')
    barras2 = ax2.bar(x + width/2, conteo_posterior, width, label=f'NDVI Posterior - {año_posterior}', color='#d7191c')

    ax2.set_xticks(x)
    ax2.set_xticklabels(etiquetas, rotation=45)
    ax2.set_title('Conteo NDVI por Categoría', fontsize=16, fontweight='bold')
    ax2.set_xlabel('Categorías NDVI', fontsize=14)
    ax2.set_ylabel('Cantidad de píxeles', fontsize=14)
    ax2.legend()

    # Mostrar valores sobre las barras
    for barra in barras1:
        altura = barra.get_height()
        ax2.annotate(f'{altura}',
                     xy=(barra.get_x() + barra.get_width() / 2, altura),
                     xytext=(0, 3),
                     textcoords="offset points",
                     ha='center', va='bottom', fontsize=10)

    for barra in barras2:
        altura = barra.get_height()
        ax2.annotate(f'{altura}',
                     xy=(barra.get_x() + barra.get_width() / 2, altura),
                     xytext=(0, 3),
                     textcoords="offset points",
                     ha='center', va='bottom', fontsize=10)

    plt.tight_layout()
    nombre_archivo = os.path.join(
        ruta_guardado,
        f'{prefijo_archivo}Comparativo_NDVI_Histograma_Barras_{año_anterior}_vs_{año_posterior}.png'
    )
    plt.savefig(nombre_archivo, dpi=300)
    plt.show()
    print(f"Imagen guardada en: {nombre_archivo}")


# Generara graficos RGB
def cargar_rgb(carpeta):
    """
    Carga las bandas B04 (Rojo), B03 (Verde), B02 (Azul) desde una carpeta dada
    y devuelve una imagen RGB normalizada.
    """
    bandas = {'red': 'B04', 'green': 'B03', 'blue': 'B02'}
    rgb = []

    for color in ['red', 'green', 'blue']:
        ruta = buscar_archivo_por_banda(carpeta, bandas[color])
        if ruta is None:
            raise FileNotFoundError(f"No se encontró la banda {bandas[color]} en {carpeta}")
        with rasterio.open(ruta) as src:
            banda = src.read(1).astype('float32')
            banda = banda / np.max(banda)  # Normalización
            rgb.append(banda)

    # Reemplaza las áreas negras (donde todas las bandas son 0) por blanco
    rgb_array = np.dstack(rgb)
    mascara_negra = np.all(rgb_array == 0, axis=-1)
    rgb_array[mascara_negra] = [1.0, 1.0, 1.0]  # Blanco
    return np.clip(rgb_array, 0, 1)

def generar_comparativo_rgb(carpeta_anterior, carpeta_posterior, año_anterior, año_posterior, prefijo_archivo):
    """
    Genera una figura con dos paneles RGB: anterior y posterior.
    """
    rgb_anterior = cargar_rgb(carpeta_anterior)
    rgb_posterior = cargar_rgb(carpeta_posterior)

    fig, axes = plt.subplots(1, 2, figsize=(18, 9))

    axes[0].imshow(rgb_anterior)
    axes[0].set_title(f"RGB - {año_anterior}", fontsize=14)
    axes[0].axis('off')

    axes[1].imshow(rgb_posterior)
    axes[1].set_title(f"RGB - {año_posterior}", fontsize=14)
    axes[1].axis('off')

    nombre_archivo = os.path.join(
        ruta_guardado,
        f'{prefijo_archivo}Comparativo_RGB_{año_anterior}_vs_{año_posterior}.png'
    )
    plt.tight_layout()
    plt.savefig(nombre_archivo, dpi=300)
    plt.show()
    print(f"Imagen RGB comparativa guardada en: {nombre_archivo}")



# ===========================================
# 6) LLAMADA A LAS FUNCIONES (EJECUCIÓN)
# ===========================================


# ------------------------------
# Visualización combinada
# ------------------------------
fig = plt.figure(figsize=(13, 10))

# Panel 1: Mapa Colombia
ax1 = plt.subplot2grid((2, 2), (0, 0), rowspan=2)
departamentos.plot(ax=ax1, color='lightgray', edgecolor='black', linewidth=0.5)
if not departamento_encontrado.empty:
    departamento_encontrado.plot(ax=ax1, color='#ffdd57', edgecolor='black', linewidth=1.5)

    # Añadir un círculo rojo para resaltar la región
    bounds = region_gdf.total_bounds
    center_x = (bounds[0] + bounds[2]) / 2
    center_y = (bounds[1] + bounds[3]) / 2
    radius = max((bounds[2] - bounds[0]), (bounds[3] - bounds[1])) / 1.5
    circle = mpatches.Circle((center_x, center_y), radius,
                           fill=False, edgecolor='red', linewidth=2.5, transform=ax1.transData)
    ax1.add_patch(circle)

ax1.set_title("Colombia - Ubicación del Departamento", fontsize=14)
ax1.axis('off')

# Panel 2: Detalle del Departamento
ax2 = plt.subplot2grid((2, 2), (0, 1))
if not departamento_encontrado.empty:
    departamento_encontrado.plot(ax=ax2, color='#ffdd57', edgecolor='black', linewidth=1.5)

    # Agregar un inset para mostrar el zoom del polígono
    # Primero, agregar un círculo rojo vacío como indicador
    bounds = region_gdf.total_bounds
    center_x = (bounds[0] + bounds[2]) / 2
    center_y = (bounds[1] + bounds[3]) / 2
    radius = max((bounds[2] - bounds[0]), (bounds[3] - bounds[1])) / 1.5
    circle = mpatches.Circle((center_x, center_y), radius,
                           fill=False, edgecolor='red', linewidth=2.5, transform=ax2.transData)
    ax2.add_patch(circle)

    # Crear un inset para mostrar el detalle del polígono
    from mpl_toolkits.axes_grid1.inset_locator import inset_axes
    axins = inset_axes(ax2, width="50%", height="50%", loc="center left")

    # Dibujar solo el polígono en el inset
    region_gdf.plot(ax=axins, color='green', edgecolor='black', linewidth=0.8)

    # Ajustar los límites del inset para mostrar solo el polígono
    axins.set_xlim(bounds[0], bounds[2])
    axins.set_ylim(bounds[1], bounds[3])
    axins.axis('off')  # Quitar ejes del inset

    ax2.set_title(f"Detalle del Departamento: {nombre_dpto}", fontsize=14)
else:
    ax2.set_title("Departamento no detectado", fontsize=14)
ax2.axis('off')

# Panel 3: NDVI Último Periodo dinámico

# Estadísticas
area_ha = calcular_area_hectareas(mascara_posterior)
ndvi_max = np.nanmax(ndvi_posterior)
ndvi_min = np.nanmin(ndvi_posterior)
ndvi_mean = np.nanmean(ndvi_posterior)

ax3 = plt.subplot2grid((2, 2), (1, 1))
norm = mcolors.Normalize(vmin=ndvi_min, vmax=ndvi_max)
cmap = plt.cm.RdYlGn
img = ax3.imshow(ndvi_posterior, cmap=cmap, norm=norm)
ax3.set_title(f"NDVI Último Periodo - {año_posterior}", fontsize=14)
ax3.axis('off')

# Barra de color horizontal
cbar = plt.colorbar(img, ax=ax3, orientation='horizontal', pad=0.15, fraction=0.040)
cbar.set_label('Valor NDVI')

# Texto en recuadro debajo de la barra de colores
ax3.text(
    0.5, -0.05,
    f"Total Área: {area_ha:.2f} ha | NDVI Máx: {ndvi_max:.2f}\n"
    f"Promedio: {ndvi_mean:.2f} | NDVI Mín: {ndvi_min:.2f}",
    transform=ax3.transAxes, fontsize=11, ha='center', va='top',
    bbox=dict(facecolor='white', edgecolor='black', boxstyle='square,pad=0.4')
)

plt.tight_layout()

# Guardar la figura
nombre_archivo = os.path.join(ruta_guardado, "Mapa_Completo_Resumen.png")
plt.savefig(nombre_archivo, dpi=300, bbox_inches='tight')
print(f"✅ Imagen guardada en: {nombre_archivo}")

plt.show()

# Genera imágenes RGB
generar_comparativo_rgb(
    carpeta_anterior, carpeta_posterior,
    año_anterior, año_posterior, prefijo_archivo)


# Gráfico NDVI (3 paneles)
generar_grafico_ndvi(
    ndvi_anterior, ndvi_posterior,
    año_anterior, año_posterior,
    mascara_anterior, mascara_posterior,
    prefijo_archivo
)

# Gráfico NDVI Categorizado
generar_grafico_ndvi_categorizado(
    ndvi_anterior, ndvi_posterior, año_anterior, año_posterior,
    diferencia_ndvi, mascara_anterior, mascara_posterior,
    grupos, etiquetas, prefijo_archivo)


# Gráfico Comparativo (Histograma vs Barras)
generar_graficos_comparativos(
    ndvi_anterior, ndvi_posterior,
    año_anterior, año_posterior, grupos,
    etiquetas, prefijo_archivo)

# Genera DF de la clasificación NDVI
df_resumen_ndvi = generar_dataframe_resumen(
    ndvi_anterior,
    ndvi_posterior,
    mascara_anterior,
    mascara_posterior,
    grupos,
    etiquetas
)

# Mostrar DataFrames resultantes
print("\n--- DataFrame de Estadísticas Generales ---")
display(df_estadisticas)
print("\n--- DataFrame de Comparación de Grupos ---")
display(df_comparativo)
print("Tabla Resumen de Áreas y Cambios por Categoría NDVI:")
display(df_resumen_ndvi)


# ===============================
# GUARDAR DATAFRAMES EN XLSX Y CSV
# ===============================
# Rutas de salida
ruta_df_estadisticas_xlsx = os.path.join(ruta_guardado, f'{prefijo_archivo}Estadisticas_Generales.xlsx')
ruta_df_comparativo_xlsx = os.path.join(ruta_guardado, f'{prefijo_archivo}Comparacion_Grupos.xlsx')
ruta_df_resumen_ndvi_xlsx = os.path.join(ruta_guardado, f'{prefijo_archivo}Resumen_Areas_NDVI.xlsx')

ruta_df_estadisticas_csv = os.path.join(ruta_guardado, f'{prefijo_archivo}Estadisticas_Generales.csv')
ruta_df_comparativo_csv = os.path.join(ruta_guardado, f'{prefijo_archivo}Comparacion_Grupos.csv')
ruta_df_resumen_ndvi_csv = os.path.join(ruta_guardado, f'{prefijo_archivo}Resumen_Areas_NDVI.csv')

# Guardar como Excel
df_estadisticas.to_excel(ruta_df_estadisticas_xlsx, index=False)
df_comparativo.to_excel(ruta_df_comparativo_xlsx, index=False)
df_resumen_ndvi.to_excel(ruta_df_resumen_ndvi_xlsx, index=False)

# Guardar como CSV
df_estadisticas.to_csv(ruta_df_estadisticas_csv, index=False)
df_comparativo.to_csv(ruta_df_comparativo_csv, index=False)
df_resumen_ndvi.to_csv(ruta_df_resumen_ndvi_csv, index=False)

print("\n✅ Archivos guardados exitosamente en formatos .xlsx y .csv:")

print("\n¡Análisis completo! Se han guardado las imágenes con el prefijo:", prefijo_archivo)
```

```python
# @title Clasificaión de coberturas Modelo SVC
# ======================
# CONFIGURACIÓN INICIAL
# ======================
import glob
import re
import matplotlib.patches as mpatches


resolucion_m = 10  # Resolución espacial de las imágenes (en metros por píxel)

# Diccionario que define las clases de cobertura del suelo según sus códigos numéricos
coberturas = {
    1: 'Bosque',
    2: 'Plantación forestal',
    3: 'Pastizales',
    4: 'Cultivos',
    5: 'sin vegetacion'
}

# Lista de colores RGB/nombres para cada clase de cobertura del suelo (para visualización)
colores = ['white', 'lime', 'green', 'red', "orange", "gray"]  # color[0] se usa para clase 0 (sin clasificar)

# Colormap personalizado usando los colores definidos
cmap = mcolors.ListedColormap(colores)

# Límites entre clases para aplicar correctamente el color a cada clase
bounds = [-0.5, 0.5, 1.5, 2.5, 3.5, 4.5, 5.5]
norm = mcolors.BoundaryNorm(bounds, cmap.N)

# ======================
# FUNCIONES AUXILIARES
# ======================

def extraer_fecha_de_ruta(ruta):
    """
    Extrae la fecha (YYYY-MM-DD) del nombre de la carpeta usando expresiones regulares.
    """
    nombre_carpeta = os.path.basename(os.path.normpath(ruta))
    match = re.search(r'(\d{4})(\d{2})(\d{2})', nombre_carpeta)
    if match:
        año, mes, dia = match.groups()
        return f"{año}-{mes}-{dia}"
    else:
        return "Fecha desconocida"

# ======================
# FUNCIÓN PRINCIPAL DE CLASIFICACIÓN
# ======================

def clasificar_imagen(directorio):
    """
    Carga las bandas espectrales desde archivos .tif, realiza la clasificación usando un modelo SVC preentrenado.
    """
    # Lista de bandas requeridas
    bandas_existentes = ["01", "02", "03", "04", "05", "06", "07", "08", "8A", "09", "11", "12"]

    # Patrón para reconocer los nombres de banda en los archivos
    patron_banda = re.compile(r'B(0[1-9]|1[12]|8A)', re.IGNORECASE)

    # Buscar todos los archivos .tif o .tiff en el directorio
    archivos_tiff = glob.glob(os.path.join(directorio, '*.tif')) + glob.glob(os.path.join(directorio, '*.tiff'))

    bandas_dict = {}

    # Leer las bandas válidas y almacenarlas
    for archivo in archivos_tiff:
        nombre = os.path.basename(archivo)
        match = patron_banda.search(nombre)
        if match:
            banda = match.group(1).upper()
            if banda in bandas_existentes:
                with rasterio.open(archivo) as src:
                    bandas_dict[banda] = src.read(1)  # Leer banda como array 2D
                    metadata = src.profile             # Guardar metadatos (resolución, proyección, etc.)

    # Ordenar bandas en el orden correcto
    bandas_ordenadas = [bandas_dict[b] for b in bandas_existentes if b in bandas_dict]

    # Verificar que todas las bandas requeridas estén presentes
    if len(bandas_ordenadas) != len(bandas_existentes):
        raise ValueError("Faltan algunas bandas requeridas para la clasificación.")

    # Apilar bandas en un array 3D
    imagen = np.stack(bandas_ordenadas, axis=-1)
    shape = imagen.shape[:2]
    imagen_flat = imagen.reshape((-1, len(bandas_existentes)))  # Aplanar para clasificación por píxel

    # Identificar píxeles inválidos (cero o NaN)
    pixeles_invalidos = np.any((imagen_flat == 0) | np.isnan(imagen_flat), axis=1)

    # Aplicar normalización solo a píxeles válidos
    imagen_flat[~pixeles_invalidos] = scaler.transform(imagen_flat[~pixeles_invalidos])

    # Inicializar array de clasificación con ceros
    clasificacion = np.full(imagen_flat.shape[0], 0)

    # Predecir clase solo para píxeles válidos
    clasificacion[~pixeles_invalidos] = modelo.predict(imagen_flat[~pixeles_invalidos])

    # Reconstruir la imagen clasificada en forma 2D
    clasificacion = clasificacion.reshape(shape)

    return clasificacion, metadata

# ======================
# FUNCIÓN DE VISUALIZACIÓN MEJORADA
# ======================
#def visualizar_comparacion_horizontal(clasif_anterior, clasif_posterior, cambio, fecha_anterior, fecha_posterior):
def visualizar_comparacion_horizontal(clasif_anterior, clasif_posterior, cambio, fecha_anterior, fecha_posterior, ruta_guardado=None, nombre_archivo="Comparativo_Coberturas.png"):

    mascara_valida = (clasif_anterior != 0) & (clasif_posterior != 0)

    valores_anterior, conteos_anterior = np.unique(clasif_anterior[mascara_valida], return_counts=True)
    valores_posterior, conteos_posterior = np.unique(clasif_posterior[mascara_valida], return_counts=True)

    area_por_pixel_ha = (resolucion_m ** 2) / 10000

    area_anterior = {
        coberturas[v]: conteos_anterior[i] * area_por_pixel_ha
        for i, v in enumerate(valores_anterior)
        if v in coberturas
    }

    area_posterior = {
        coberturas[v]: conteos_posterior[i] * area_por_pixel_ha
        for i, v in enumerate(valores_posterior)
        if v in coberturas
    }

    fig, axs = plt.subplots(1, 3, figsize=(24, 10))
    fontsize_title = 14
    fontsize_stats = 10

    def formatear_area(area_dict):
        total = sum(area_dict.values())
        texto = ""
        for k, v in area_dict.items():
            pct = (v / total) * 100 if total > 0 else 0
            texto += f"{k}: {v:.2f} ha ({pct:.1f}%)\n"
        return texto.strip()

    # Clasificación anterior
    axs[0].imshow(clasif_anterior, cmap=cmap, norm=norm)
    axs[0].set_title(f"Clasificación - {fecha_anterior}", fontsize=fontsize_title, fontweight='bold')
    axs[0].axis("off")
    axs[0].text(0.5, -0.08, formatear_area(area_anterior), transform=axs[0].transAxes,
                fontsize=fontsize_stats, va='top', ha='center',
                bbox=dict(facecolor='white', edgecolor='black'))

    # Clasificación posterior
    axs[1].imshow(clasif_posterior, cmap=cmap, norm=norm)
    axs[1].set_title(f"Clasificación - {fecha_posterior}", fontsize=fontsize_title, fontweight='bold')
    axs[1].axis("off")
    axs[1].text(0.5, -0.08, formatear_area(area_posterior), transform=axs[1].transAxes,
                fontsize=fontsize_stats, va='top', ha='center',
                bbox=dict(facecolor='white', edgecolor='black'))

    # Imagen de cambio
    cambio_rgb = np.full((*cambio.shape, 3), [255, 255, 255], dtype=np.uint8)
    cambio_rgb[(cambio == 0) & mascara_valida] = [0, 0, 255]
    cambio_rgb[(cambio == 1) & mascara_valida] = [200, 0, 0]

    axs[2].imshow(cambio_rgb)
    axs[2].set_title(f"Cambio de coberturas  ({fecha_anterior} - {fecha_posterior})", fontsize=fontsize_title, fontweight='bold')
    axs[2].axis("off")

    pixeles_cambio = np.count_nonzero((cambio == 1) & mascara_valida)
    pixeles_sin_cambio = np.count_nonzero((cambio == 0) & mascara_valida)
    ha_cambio = pixeles_cambio * area_por_pixel_ha
    ha_sin_cambio = pixeles_sin_cambio * area_por_pixel_ha
    total = ha_cambio + ha_sin_cambio
    pct_cambio = (ha_cambio / total) * 100 if total > 0 else 0
    pct_sin_cambio = (ha_sin_cambio / total) * 100 if total > 0 else 0

    texto_cambio = (
        f"Sin cambio: {ha_sin_cambio:.2f} ha ({pct_sin_cambio:.1f}%)\n"
        f"Cambio en la cobertura: {ha_cambio:.2f} ha ({pct_cambio:.1f}%)"
    )

    axs[2].text(0.5, -0.08, texto_cambio, transform=axs[2].transAxes,
                fontsize=fontsize_stats, va='top', ha='center',
                bbox=dict(facecolor='white', edgecolor='black'))

    # Leyenda coberturas centrada abajo (horizontal)
    patches_cobertura = [mpatches.Patch(color=colores[i], label=coberturas[i]) for i in coberturas]
    fig.legend(handles=patches_cobertura,
               loc='upper center',
               bbox_to_anchor=(0.4, -0.01),
               ncol=len(coberturas),
               fontsize=fontsize_stats,
               frameon=True,
               fancybox=True,
               edgecolor='gray',
               facecolor='white',
               framealpha=0.8 ) # Transparencia del fondo


    # Leyenda cambio centrada abajo (horizontal)
    patch_sin_cambio = mpatches.Patch(color='blue', label='Sin cambio')
    patch_con_cambio = mpatches.Patch(color='red', label='Cambio en la cobertura')
    fig.legend(handles=[patch_sin_cambio, patch_con_cambio],
               loc='upper center',
               bbox_to_anchor=(0.79, -0.01),
               ncol=2,
               fontsize=fontsize_stats,
               frameon=True,
               fancybox=True,
               edgecolor='gray',
               facecolor='white',
               framealpha=0.8)

    plt.subplots_adjust(bottom=0.15)
    #plt.show()
    if ruta_guardado:
        ruta_final = os.path.join(ruta_guardado, nombre_archivo)
        plt.savefig(ruta_final, dpi=300, bbox_inches='tight')
        print(f"✅ Imagen de cambio de cobertura guardada en: {ruta_final}")
    else:
        plt.show()

    return area_anterior, area_posterior # Added this line to return the area dictionaries

    # ======================
# EJECUCIÓN DEL ANÁLISIS
# ======================

# Extraer las fechas de las carpetas con las imágenes
fecha_anterior = extraer_fecha_de_ruta(carpeta_anterior)
fecha_posterior = extraer_fecha_de_ruta(carpeta_posterior)

# Clasificar ambas imágenes
clasif_anterior, metadata = clasificar_imagen(carpeta_anterior)
clasif_posterior, _ = clasificar_imagen(carpeta_posterior)

# Generar mapa de cambio binario (1 si hay cambio, 0 si se mantiene igual)
cambio_binario = (clasif_anterior != clasif_posterior).astype(np.uint8)

# Visualizar las dos clasificaciones y el cambio
area_anterior,  area_posterior = visualizar_comparacion_horizontal(
    clasif_anterior,
    clasif_posterior,
    cambio_binario,
    fecha_anterior,
    fecha_posterior,
    ruta_guardado=ruta_guardado,  # tu variable de carpeta destino
    nombre_archivo="Comparativo_Coberturas.png"
)


claves_cobertura = list(coberturas.values())

df_comparacion = pd.DataFrame({
    "Cobertura": claves_cobertura,
    f"Área ({fecha_anterior}) (Ha)": [area_anterior.get(cob, 0) for cob in claves_cobertura],
    f"Área ({fecha_posterior}) (Ha)": [area_posterior.get(cob, 0) for cob in claves_cobertura],
})

df_comparacion.to_csv(os.path.join(ruta_guardado, 'Comparacion_Coberturas.csv'), index=False)
print(df_comparacion)
```
